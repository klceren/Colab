```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Veriyi simüle ediyoruz
data = np.random.randint(0, 255, (1000, 1200)).tolist()
labels = [1 if i < 500 else 0 for i in range(1000)]
dataset = {"data": data, "labels": labels}

with open('planesnet.json', 'w') as f:
    json.dump(dataset, f)

X = np.array(dataset['data']).astype('uint8')
y = np.array(dataset['labels']).astype('uint8')
print("Veri hazır!")
```

```python
plt.figure(figsize=(10, 4))
for i in range(5):
    plt.subplot(1, 5, i+1)
    img = X[i].reshape(3, 400).T.reshape(20, 20, 3)
    plt.imshow(img)
    plt.title("Ucak" if y[i] == 1 else "Degil")
    plt.axis('off')
plt.show()
```

```python
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

sns.countplot(x=y)
plt.title("Sınıf Dağılımı")
plt.show()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = LogisticRegression(max_iter=1000).fit(X_train, y_train)
print(f"Başarı Oranı: %{model.score(X_test, y_test)*100}")
```
